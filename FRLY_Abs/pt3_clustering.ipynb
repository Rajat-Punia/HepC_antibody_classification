### Script summary (pairwise rotation-angle between Fv regions)

- **Goal:** Quantify orientation differences between antibody **Fv** regions (H≤150 + L≤150) across E2-aligned complexes.
- **Input:**  
  - `input.csv` with `pdb_id, antigen_chain_id, heavy_chain_id, light_chain_id, antibody`  
  - Aligned PDBs in `aligned_structures_E2/`
- **Alignment & angle:**  
  - Runs `tmtools.tm_align` on **Fv coords** for each pair (j vs i).  
  - Gets rotation matrix `R` from the alignment; computes **angle** `θ = arccos((trace(R)−1)/2)` (degrees).  
  - Stores θ in a **symmetric distance matrix**.
- **Output:**  
  - Saves labeled angle matrix as **`angle_FRLY_Abs.csv`**.  


```python
import pandas as pd
import numpy as np
from numpy.linalg import norm
import matplotlib.pyplot as plt
from Bio.PDB import PDBParser
from tmtools import tm_align
from tmtools.io import get_structure, get_residue_data

try:
    from Bio.PDB.Polypeptide import protein_letters_3to1
except ImportError:
    from Bio.PDB import protein_letters_3to1
    
# Load the CSV file
csv_file = 'input.csv'
df = pd.read_csv(csv_file, header=None, names=['pdb_id', 'antigen_chain_id', 'heavy_chain_id', 'light_chain_id','antibody'])

parser = PDBParser(QUIET=True)

# Function to extract antigen data from a structure
def extract_Ag_data(structure, chain_id):
    chain = structure[0][chain_id]
    coords = []
    seq = ''
    for res in chain:
        if 'CA' in res.child_dict:  
            coords.append(res['CA'].coord)
            if res.resname == 'HSD' or res.resname == 'HSP':
                seq += 'H' 
            else:
                seq += protein_letters_3to1[res.resname]  
    return np.array(coords), seq

# Function to extract Fv data from a structure
def extract_fv_data(structure, chain_id):
    chain = structure[0][chain_id]
    coords = []
    seq = ''
    for res in chain:
        if res.id[1] <= 150: 
            if 'CA' in res.child_dict: 
                coords.append(res['CA'].coord)
                if res.resname == 'HSD' or res.resname == 'HSP':
                    seq += 'H' 
                else:
                    seq += protein_letters_3to1[res.resname]  
    return np.array(coords), seq

# Initialize distance matrix
n = len(df)
distance_matrix = np.zeros((n, n))

# Iterate through pairs
for i in range(len(df)):
    struct_path = f"aligned_structures_E2/aligned_{df.iloc[i]['pdb_id']}_{df.iloc[i]['antigen_chain_id']}_{df.iloc[i]['heavy_chain_id']}_{df.iloc[i]['light_chain_id']}.pdb"
    structure = get_structure(struct_path)
    coords_Ag, seq_Ag = extract_Ag_data(structure, df.iloc[i]['antigen_chain_id'])

    coords_h, seq_h = extract_fv_data(structure, df.iloc[i]['heavy_chain_id'])
    coords_l, seq_l = extract_fv_data(structure, df.iloc[i]['light_chain_id'])
    coords_fv = np.concatenate((coords_h, coords_l), axis=0)
    seq_fv = seq_h + seq_l

    for j in range(i + 1, len(df)):
        other_struct_path = f"aligned_structures_E2/aligned_{df.iloc[j]['pdb_id']}_{df.iloc[j]['antigen_chain_id']}_{df.iloc[j]['heavy_chain_id']}_{df.iloc[j]['light_chain_id']}.pdb"
        other_structure = get_structure(other_struct_path)
        other_coords_Ag, other_seq_Ag = extract_Ag_data(other_structure, df.iloc[j]['antigen_chain_id'])

        other_coords_h, other_seq_h = extract_fv_data(other_structure, df.iloc[j]['heavy_chain_id'])
        other_coords_l, other_seq_l = extract_fv_data(other_structure, df.iloc[j]['light_chain_id'])
        other_coords_fv = np.concatenate((other_coords_h, other_coords_l), axis=0)
        other_seq_fv = other_seq_h + other_seq_l


        # Align using tmtools 
        result = tm_align(other_coords_fv, coords_fv, other_seq_fv, seq_fv)
        
        
        # Store the transformation matrix and vector
        rotation_matrix = np.array(result.u, dtype=float)
        theta = np.arccos((np.trace(rotation_matrix) - 1) / 2.0)
        angle = np.degrees(theta)
        
        distance_matrix[i, j] = angle
        distance_matrix[j, i] = angle  # Symmetric

labels = [f"{row['antibody']}" for index, row in df.iterrows()]

distance_df = pd.DataFrame(distance_matrix, index=labels, columns=labels)
distance_df.to_csv('angle_FRLY_Abs.csv', index=True)

# Display the DataFrame
from IPython.display import display
formatted_df = distance_df.style.format("{:.1f}")
display(formatted_df)
```

### Histogram of antibody approach angles

### DBSCAN clustering 

- **Goal:** Cluster antibodies by pairwise **approach angles** (`distance_matrix`) 
- **Clustering:** Runs **DBSCAN** (`eps=35`, `metric='precomputed'`, `min_samples=1`) to assign a cluster ID to each antibody.


```python
from sklearn.cluster import DBSCAN
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Apply DBSCAN
eps_value = 35
dbscan = DBSCAN(eps=eps_value, min_samples=1, metric='precomputed')  
clusters = dbscan.fit_predict(distance_matrix)

# Create a DataFrame 
pdb_ids = df['antibody'].tolist()
cluster_data = pd.DataFrame({'pdb_id': pdb_ids, 'cluster': clusters, 'index': range(len(pdb_ids))})

cluster_sizes = cluster_data['cluster'].value_counts().reset_index()
cluster_sizes.columns = ['cluster', 'size']

cluster_sizes_sorted = cluster_sizes.sort_values(by='size', ascending=False)

custom_cluster_order = {0: 3, 1: 0, 2: 4, 3: 1, 4: 2} 

cluster_data['cluster_order'] = cluster_data['cluster'].map(custom_cluster_order)

sorted_cluster_data = cluster_data.sort_values(by=['cluster_order', 'index'])

# Extract the ordered indices and pdb_ids for labeling
sorted_indices = sorted_cluster_data['index'].tolist()
sorted_pdb_ids = sorted_cluster_data['pdb_id'].tolist()

# Reorder the distance matrix according to the sorted indices
sorted_distance_matrix = distance_matrix[np.ix_(sorted_indices, sorted_indices)]

# Custom RGB colors
custom_colors = {
    0: (0.6, 0.1, 0.6), # For cluster 0
    1: (0.0, 0.5, 1),   # For cluster 1
    2: (0.6, 0.5, 1), # For cluster 2
    3: (1.0, 0.2, 0.2), # For cluster 2
    4: (1, 0.5, 0), # For cluster 2
}

fallback_color = (0.5, 0.5, 0.5)  # Gray

# Plot the heatmap 
plt.figure(figsize=(12, 9))
ax = sns.heatmap(sorted_distance_matrix, cmap='RdYlBu_r', alpha=0.8, xticklabels=sorted_pdb_ids, yticklabels=sorted_pdb_ids)

plt.xlabel('Antibody (PDB ID)', fontsize=14)
plt.ylabel('Antibody (PDB ID)', fontsize=14)

ax.xaxis.set_ticks_position('top')
ax.xaxis.set_label_position('top')

for i, tick_label in enumerate(ax.get_xticklabels()):
    cluster_id = sorted_cluster_data.iloc[i]['cluster']
    color = custom_colors.get(cluster_id, fallback_color)
    tick_label.set_color(color)

for i, tick_label in enumerate(ax.get_yticklabels()):
    cluster_id = sorted_cluster_data.iloc[i]['cluster']
    color = custom_colors.get(cluster_id, fallback_color)
    tick_label.set_color(color)
    

ax.set_xticklabels(ax.get_xticklabels(), rotation=60, ha='left', fontsize=10)
ax.set_yticklabels(ax.get_yticklabels(), rotation=0, ha='right', fontsize=10)

colorbar = ax.collections[0].colorbar
colorbar.ax.tick_params(labelsize=12)

plt.show()
```
